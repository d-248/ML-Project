```python
############################################################
# ITERATION FOR HYPERPARAMETER SWEEP
# EXTRAPOLATION SPLIT + FIXED TEST SET
############################################################

# ===================== IMPORT LIBRARIES =====================
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import backend as K
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from rdkit import Chem
from rdkit.Chem import AllChem
from itertools import product

# ===================== LOAD DATA =====================
data = pd.read_csv("qm133885_properties.csv")

smiles = data["SMILES"]
y = data["H"].values

# ===================== SMILES → Morgan Fingerprint =====================
def smiles_to_fingerprint(sm):
    mol = Chem.MolFromSmiles(sm)
    if mol is None:
        return np.zeros(1024)

    return np.array(
        AllChem.GetMorganFingerprintAsBitVect(
            mol,
            radius=3,
            nBits=1024
        ),
        dtype=np.float32
    )

# ===============================
# CREATE FINGERPRINT MATRIX
# ===============================
print("Generating fingerprints...")

X = np.array(
    [smiles_to_fingerprint(s) for s in smiles],
    dtype=np.float32
)

print("Fingerprint shape:", X.shape)

# ===============================
# EXTRAPOLATION SPLIT
# ===============================
train_mask = (y >= -460) & (y <= -360)
test_mask = ~train_mask

X_train_full = X[train_mask]
y_train_full = y[train_mask]

X_test_all = X[test_mask]
y_test_all = y[test_mask]

print("\nTrain H range:",
      y_train_full.min(),
      y_train_full.max())

print("Test H range :",
      y_test_all.min(),
      y_test_all.max())

print("Available train samples:",
      len(y_train_full))

print("Available test samples:",
      len(y_test_all))

# ===============================
# FIX TEST SET TO 24,869 SAMPLES
# ===============================
np.random.seed(42)

test_indices = np.random.choice(
    len(X_test_all),
    size=24869,
    replace=False
)

X_test = X_test_all[test_indices]
y_test = y_test_all[test_indices]

print("Fixed Test Samples:",
      len(X_test))

# ===============================
# TRAIN / VALIDATION SPLIT
# ===============================
X_train, X_val, y_train, y_val = train_test_split(
    X_train_full,
    y_train_full,
    test_size=0.20,
    random_state=42
)

print("\nTrain samples:",
      len(y_train))

print("Validation samples:",
      len(y_val))

print("Fixed test samples:",
      len(y_test))

# ===============================
# SCALE DATA
# ===============================
x_scaler = StandardScaler()
y_scaler = StandardScaler()

X_train_scaled = x_scaler.fit_transform(X_train)
X_val_scaled = x_scaler.transform(X_val)
X_test_scaled = x_scaler.transform(X_test)

y_train_scaled = y_scaler.fit_transform(
    y_train.reshape(-1, 1)
).flatten()

y_val_scaled = y_scaler.transform(
    y_val.reshape(-1, 1)
).flatten()

y_test_scaled = y_scaler.transform(
    y_test.reshape(-1, 1)
).flatten()

# ===============================
# MODEL FUNCTION
# ===============================
def create_model(
    neurons1=32,
    neurons2=8,
    neurons3=8,
    lr=0.001,
    activation="relu"
):

    model = tf.keras.Sequential([
        tf.keras.layers.Input(shape=(1024,)),
        tf.keras.layers.Dense(
            neurons1,
            activation=activation
        ),
        tf.keras.layers.Dense(
            neurons2,
            activation=activation
        ),
        tf.keras.layers.Dense(
            neurons3,
            activation=activation
        ),
        tf.keras.layers.Dense(1)
    ])

    model.compile(
        optimizer=Adam(learning_rate=lr),
        loss="mse",
        metrics=["mae"]
    )

    return model

# ===============================
# HYPERPARAMETER GRID
# ===============================
param_grid = {
    "neurons1": [8,16,32,64,128,256,512,1024],
    "neurons2": [8,16,32,64,128,256,512,1024],
    "neurons3": [8,16,32,64,128,256,512,1024],
    "lr": [0.001],
    "activation": ["relu"]
}

# ===============================
# HYPERPARAMETER SWEEP
# ===============================
results = []

all_params = list(
    product(
        param_grid["neurons1"],
        param_grid["neurons2"],
        param_grid["neurons3"],
        param_grid["lr"],
        param_grid["activation"]
    )
)

print("\nTotal combinations:",
      len(all_params))

for i, (
    neurons1,
    neurons2,
    neurons3,
    lr,
    activation
) in enumerate(all_params, start=1):

    print(
        f"\n[{i}/{len(all_params)}] "
        f"n1={neurons1}, "
        f"n2={neurons2}, "
        f"n3={neurons3}, "
        f"lr={lr}"
    )

    K.clear_session()

    model = create_model(
        neurons1=neurons1,
        neurons2=neurons2,
        neurons3=neurons3,
        lr=lr,
        activation=activation
    )

    history = model.fit(
        X_train_scaled,
        y_train_scaled,
        validation_data=(
            X_val_scaled,
            y_val_scaled
        ),
        epochs=200,
        batch_size=100,
        verbose=0
    )

    # ===============================
    # PREDICTIONS
    # ===============================
    y_train_pred = y_scaler.inverse_transform(
        model.predict(
            X_train_scaled,
            verbose=0
        )
    ).flatten()

    y_val_pred = y_scaler.inverse_transform(
        model.predict(
            X_val_scaled,
            verbose=0
        )
    ).flatten()

    y_test_pred = y_scaler.inverse_transform(
        model.predict(
            X_test_scaled,
            verbose=0
        )
    ).flatten()

    # ===============================
    # R² SCORES
    # ===============================
    r2_train = r2_score(
        y_train,
        y_train_pred
    )

    r2_val = r2_score(
        y_val,
        y_val_pred
    )

    r2_test = r2_score(
        y_test,
        y_test_pred
    )

    results.append({
        "neurons1": neurons1,
        "neurons2": neurons2,
        "neurons3": neurons3,
        "lr": lr,
        "activation": activation,
        "r2_train": r2_train,
        "r2_val": r2_val,
        "r2_test": r2_test
    })

    print(
        f"Train R²={r2_train:.4f} | "
        f"Val R²={r2_val:.4f} | "
        f"Test R²={r2_test:.4f}"
    )

# ===============================
# BEST MODEL (SELECTED BY VAL)
# ===============================
best_result = max(
    results,
    key=lambda x: x["r2_val"]
)

print("\n================ BEST RESULT ================")
print("Best Validation R² :", best_result["r2_val"])
print("Train R²           :", best_result["r2_train"])
print("Test R²            :", best_result["r2_test"])

print("\nBest Parameters")
print("neurons1   :", best_result["neurons1"])
print("neurons2   :", best_result["neurons2"])
print("neurons3   :", best_result["neurons3"])
print("lr         :", best_result["lr"])
print("activation :", best_result["activation"])
```
